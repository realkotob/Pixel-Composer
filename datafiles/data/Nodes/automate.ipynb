```python
import os

_dir = "D:/Project/MakhamDev/LTS-PixelComposer/PixelComposer/datafiles/data/Nodes/Internal"
_dirMap = {}

for folder in os.listdir(_dir):
    _p = os.path.join(_dir, folder)
    if not os.path.isdir(_p):
        continue

    for file in os.listdir(_p):
        _fp = os.path.join(_p, file)

        _dirMap[file.lower()] = _fp

print(_dirMap)
    
```

```python
_imgDir = "D:/Project/MakhamDev/LTS-PixelComposer/PixelComposer/datafiles/data/Nodes/Tooltip"

for file in os.listdir(_imgDir):
    _fp = os.path.join(_imgDir, file)

    fkey = file.lower().replace(".png", "")
    if fkey not in _dirMap:
        print("File not found: ", fkey)
        continue
    
    _tarDir = _dirMap[fkey]
    _nf = os.path.join(_tarDir, "tooltip_spr.png")
    os.rename(_fp, _nf)
    print("Move: ", _fp, " -> ", _nf)
        
```
